```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import pandas as pd
import time

# Initialize geolocator
geolocator = Nominatim(user_agent="p.argento@studenti.unipi.it")
chicago_bbox = [41.644531,42.02304,-87.940101,-87.524081]

def build_address(data):
    address = {
      "city": "Chicago",
      "country": "USA",
      "house_number": str(data["STREET_NO"]),
      "road": str(data["STREET_NAME"]),
      "boundingbox": chicago_bbox
    }
    return address

# Function to get coordinates
def get_coordinates(address):
    try:
        location = geolocator.geocode(address, timeout=10)
        time.sleep(1)  # Respect Nominatim's policy of 1 request per second
        if location:
            return location.latitude, location.longitude
        else:
            print("not a location")
            return None, None
    except GeocoderTimedOut:
        print("timeout")
        return None, None
```

```python
data = missing_beats.iloc[2]
address = build_address(data)
print(address)
coord = get_coordinates(address)
print(coord)
missing_beats.loc[missing_beats.index[2], ["LATITUDE", "LONGITUDE"]] = coord
```

```python
# cheche the bbox

import folium

# Define the bounding box coordinates
bbox_coords = [(-87.940101, 41.644531), (-87.524081, 42.02304)]

# Define the point to add to the map
point_coords = (41.8755616, -87.6244212)

# Calculate the map center from the bounding box
map_center = [(bbox_coords[0][1] + bbox_coords[1][1]) / 2, (bbox_coords[0][0] + bbox_coords[1][0]) / 2]

# Create a folium map centered on the bounding box
map_folium = folium.Map(location=map_center, zoom_start=12)

# Add the bounding box as a rectangle
folium.Rectangle(
    bounds=[(bbox_coords[0][1], bbox_coords[0][0]), (bbox_coords[1][1], bbox_coords[1][0])],
    color="blue",
    fill=True,
    fill_opacity=0.2
).add_to(map_folium)

# Add the point to the map
folium.Marker(
    location=[point_coords[0], point_coords[1]],
    popup="Point: (41.8755616, -87.6244212)",
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(map_folium)

# Display the map
map_folium
```
